```python
import numpy as np
import matplotlib.pyplot as plt

# utworzenie funkcji przy pomocy konstrukcji lambda
fx = lambda x: x**2 + 5

# utworzenie danych z odpowiednich zakres√≥w
x_1 = np.linspace(-1, 1, 1000)
x_2 = np.linspace(-6, 6, 1000)
x_3 = np.linspace(0, 5, 1000)
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=[4,8])
ax[0].plot(x_1, fx(x_1))
ax[0].set(title="Wykres <-1,1>")
ax[1].plot(x_2, fx(x_2))
ax[1].set(title="Wykres <-6,6>")
ax[2].plot(x_3, fx(x_3))
ax[2].set(title="Wykres <0,5>")
plt.tight_layout()
plt.show()
```
